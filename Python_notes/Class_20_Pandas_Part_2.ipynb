# Slicing

```python
1. Normal slicing - It does slicing only on rows


syntax 1

dataframe[start_index : stop_index : step]
- Start value is inclusive
- stop value is exclusive


syntax 2

dataframe [start_row : stop_row : step]
- Start value is inclusive
- stop value is inclusive
```

```python
# 3,4,5,6

df[3:7]
```

```python
# 1,3,5,7,9

df[1:10:2]
```

```python
df[::5]
```

```python
# 

new_df = df[:6]
```

# Loc method


- does slicing both on rows and columns

syntax

dataframe.loc[row_name_start : row_name_stop: row_step , column_start:stop:step]

- Normal slicing is not possible

```python
new_df['row_2':'row_5']
```

```python
new_df.loc['row_2':'row_5']
```

```python
new_df.loc[3:5]
```

```python
df[2:4]
```

```python
df.loc[2:4]
```

```python
df.loc[3:5 , 'tripduration':'bikeid']
```

```python
new_df.loc[::2 , ::2]
```

```python
df.loc[0:10 , 'start station id' : 'end station name']
```

```python
df.loc[0:5 , 3:6]
```

```python
# Extract all the data where usertype == Subscriber

df.loc[df['usertype'] == 'Subscriber']
```

```python
# (usertype == Subscriber) & (tripduration > 300)
df.loc[(df['usertype'] == 'Subscriber') & (df['tripduration']> 300)]
```

```python
df.loc[(df['usertype'] == 'Subscriber') | (df['tripduration']> 300)]
```
