# Basic EDA (Exploratory data analysis)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df
```

```python
# alive , pclass , who , adult_male , embarked

del -----> alive , pclass
drop ---> who , adult_male , embarked
```

```python
df.drop(['who','adult_male','embarked'], axis=1 , inplace = True)
del [df['pclass'] , df['alive']]


# df[['pclass' , 'alive']]
# df.drop(cloumns = ['who','adult_male','embarked'] , inplace = True)
```

```python
# age ---> mean
# deck ---> mode()
# embark_town ---> mode()

df.fillna({'age': df['age'].mean() ,
          'deck': df['deck'].mode()[0] , 
          'embark_town': df['embark_town'].mode()[0]} , inplace = True)
```

```python
df_first = df.groupby('class').get_group('First')
```

```python
df_first.loc[df_first['survived'] == 1]
```

```python
elist_1 = []
elist_2 = []

for i in df['class'].unique():
    df_trial = df.groupby('class').get_group(i)
    print(f"{i} ---- {len(df_trial.loc[df_trial['survived'] == 1])}")
    elist_1.append(i)
    elist_2.append(len(df_trial.loc[df_trial['survived'] == 1]))
```

```python
plt.bar(elist_1 , elist_2)
```
