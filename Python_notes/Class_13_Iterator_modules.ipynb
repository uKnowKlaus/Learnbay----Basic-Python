# Different ways of importing a module

```python
# first way

print(dir() , end = '\n\n')

import math

print(dir())
```

```python
print(dir(math))
```

```python
math.sqrt(9)
```

```python
# second way

print(dir() , end = '\n\n')

from math import *

print(dir())
```

```python
sqrt(13)
```

```python
# Third way

print(dir() , end = '\n\n')

from math import sqrt

print(dir())
```

```python
sqrt(9)
```

```python
# Fourth method - Aliasing

print(dir() , end = '\n\n')

import math as mt

print(dir())
```

```python
print(dir(mt))
```

```python
mt.sqrt(9)
```
