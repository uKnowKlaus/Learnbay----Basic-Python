# Iterating over a data frame

1. itertuples
2. iterrows
3. iteritems

```python
space_org = {'Organizations_name' : ["CNES", "ISRO", "NASA", "JAXA", "CNSA"],
                   'Headquarters' : ["Paris", "Bangalore", "Washington",
                                     "Tokyo", "Beijing"],
                   'Country_name' : ["France", "India", "USA", "Japan", "China"],
                 'Formation_year' : [1961, 1969, 1958, 2003, 1993]
             }
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame(space_org)
df
```

```python
# Iter tuples


for i in df:
    print(i) 
```

```python
# Iter tuples


for i in df.itertuples():
    if i.Organizations_name == 'ISRO':
        print(i.Country_name)
```

```python
df
```

```python
# Iter rows

for i,j in df.iterrows():
    print(j , end = '\n\n\n')
```

```python
# Iter rows

for i,j in df.iterrows():
    print(j['Organizations_name'])
```

```python
# Iter rows

for i,j in df.iterrows():
    if j[0] == 'ISRO':
        print(j['Country_name'])
```

# Merge

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","orlando"],
    "humidity": [65,68,75],
})
df2
```

```python
pd.concat([df1 , df2])
```

```python
pd.merge(df1 , df2)
```

```python
df1 = pd.DataFrame({
    "city": ["new york","chicago","orlando", "baltimore"],
    "temperature": [21,14,35, 38],
})
df1
```

```python
df2 = pd.DataFrame({
    "city": ["chicago","new york","san diego"],
    "humidity": [65,68,71],
})
df2
```

```python
pd.merge(df1 ,df2)
```

```python
pd.merge(df1 ,df2, how = 'outer')
```
