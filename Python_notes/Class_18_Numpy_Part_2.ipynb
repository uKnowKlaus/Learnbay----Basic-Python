```python
def apartmentHunting(blocks, reqs):
    req_dist = {}
    min_index = 0
    shortest = None
    farthest = 0
    for i in range(len(blocks)):
        #print(f'Index {i}:  --> ', end='')
        meet_req = False
        exact_match = True
        req_dist_iter = {}
        for j in reqs:
            req_dist_iter.setdefault(j,None)
            if blocks[i][j]:
                req_dist_iter[j] = 0
            else:
                for ii in range(i-1, -1,-1):
                    if blocks[ii][j]:
                        min_dist = i - ii
                        if req_dist_iter[j] is None:
                            req_dist_iter[j] = min_dist
                for ii in range(i+1, len(blocks)):
                    if blocks[ii][j]:
                        min_dist = ii - i
                        if req_dist_iter[j] is None:
                            req_dist_iter[j]= min_dist
        farthest = max(req_dist_iter.values())
        req_dist_iter.setdefault("farthest",farthest)
        req_dist.setdefault(i,req_dist_iter)
        if shortest is None:
            shortest = farthest
            min_index = i
        elif shortest > farthest:
            shortest = farthest
            min_index = i
        #print(f'Distances at index {i} : {req_dist[i]}  \n')
    print(f"\tAt index {min_index} the farthest you'd have to walk to reach a {', a '.join(reqs)} is {shortest} block(s). \n \tAt any other index, you'd have to walk farther \n")

blks = [
    { "gym": False, "school": True, "store": False, },
    { "gym": True, "school": False, "store": False, },
    { "gym": True, "school": True, "store": False, },
    { "gym": False, "school": True, "store": False, },
    { "gym": False, "school": True, "store": True, } ]

rqs = ["gym", "school", "store"]
print(f'{rqs}: ')
apartmentHunting(blks, rqs)

rqs = ["gym", "school"]
print(f'{rqs}: ')
apartmentHunting(blks, rqs)

rqs = ["store"]
print(f'{rqs}: ')
apartmentHunting(blks, rqs)

rqs = ["gym", "store"]
print(f'{rqs}: ')
apartmentHunting(blks, rqs)

rqs = ["school", "store"]
print(f'{rqs}: ')
apartmentHunting(blks, rqs)

rqs = ["school"]
print(f'{rqs}: ')
apartmentHunting(blks, rqs)
```

```python
def apartmentHunting(blocks, reqs):
    result = [float('-inf') for _ in range(len(blocks))]
    
    for currentIdx in range(len(blocks)):   
        for req in reqs: 
            step = float('inf')   
            # compare from end to start
            for compareIdx in range(currentIdx - 1, -1, -1):   
                if blocks[compareIdx][req]:                    
                    step = min(step, currentIdx - compareIdx)    
            # compare from start to end
            for compareIdx in range(currentIdx, len(blocks)):   
                if blocks[compareIdx][req]:
                    step = min(step, compareIdx - currentIdx)    
            result[currentIdx] = max(result[currentIdx], step)      
    return result.index(min(result))

# result[0]=1
# result[1]=1
# result[2]=0
a= apartmentHunting(blocks, reqs)
print(a)
```
