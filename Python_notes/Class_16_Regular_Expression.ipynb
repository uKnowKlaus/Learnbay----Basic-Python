```python
class Over_limit(Exception):
    def __init__(self , msg):
        self.msg = msg
        
class Pan_error(Exception):
    def __init__(self , msg):
        self.msg = msg
        
class Terminate_error(Exception):
    def __init__(self , msg):
        self.msg = msg
        
        
def amount_deposit(amt):
    try:
        if amt <= 50000:
            return "Amount deposited"
        elif amt >50000 and amt <=60000:
            raise Pan_error ('The amount is out of the range please enter the pan : ')
        else:
            raise Over_limit('The amount is greater than 60000 please check the values : ')
    except Over_limit as e:
        print(e)
        a = input('Do you want to retry it? Y/N : ')
        if a == 'Y':
            amt_1 = eval(input('Enter the amount : '))  
            print(amount_deposit(amt_1))
        elif a == "N":
            raise Terminate_error('come back later')
        else:
            return ('Please enter an valid option')
        
def sumbit_pan(pan):
    if len(pan) == 10:
        return 'Amount deposited'
    else:
        raise Terminate_error('Please try after sometime')

        
amt = eval(input('Enter the amount : '))        
try:
    print(amount_deposit(amt))
except Pan_error as e:
    print(e)
    try:
        print("You'll need to enter the pan for further processing" )
        pan = input('Enter the Pan number : ')
        print(sumbit_pan(pan))
    except Terminate_error as e:
        print(e)
except Terminate_error as e:
    print(e)
```

# Regular expressions

```python
'''
.       - Any Character Except New Line
\d      - Digit (0-9)
\D      - Not a Digit (0-9)
\w      - Word Character (a-z, A-Z, 0-9, _)
\W      - Not a Word Character
\s      - Whitespace (space, tab, newline)
\S      - Not Whitespace (space, tab, newline)

\b      - Word Boundary
\B      - Not a Word Boundary
^       - Beginning of a String
$       - End of a String

[]      - Matches Characters in brackets
[^ ]    - Matches Characters NOT in brackets
|       - Either Or
( )     - Group

Quantifiers:
*       - 0 or More
+       - 1 or More
?       - 0 or One
{3}     - Exact Number
{3,4}   - Range of Numbers (Minimum, Maximum)
'''
```
